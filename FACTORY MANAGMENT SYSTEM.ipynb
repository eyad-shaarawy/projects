```python
# Machine Class 
class Machine:
    def __init__(self, name, production_line=None):
        self.name = name
        self.status = "Idle"  # Idle / Running / Maintenance
        self.current_product = None
        self.production_line = production_line

    def start(self, product):
        if self.status == "Idle":
            self.status = "Running"
            self.current_product = product
            print(f"{self.name} started producing {product.name}")
        elif self.status == "Running":
            print(f"{self.name} is already running {self.current_product.name}!")
        else:
            print(f"{self.name} cannot start, status: {self.status}")

    def stop(self):
        if self.status == "Running":
            print(f"{self.name} stopped producing {self.current_product.name}")
            self.status = "Idle"
            self.current_product = None

    def maintain(self):
        self.status = "Maintenance"
        print(f"{self.name} is under maintenance")


#  Product Class 
class Product:
    def __init__(self, name):
        self.name = name
        self.quantity = 0

    def produce(self, amount):
        self.quantity += amount
        print(f"{amount} units of {self.name} produced. Total: {self.quantity}")


#  Worker Classes 
class Worker:
    def __init__(self, name):
        self.name = name


class RegularWorker(Worker):
    def __init__(self, name):
        super().__init__(name)
        self.assigned_machine = None

    def assign_machine(self, machine):
        if not machine.production_line:
            print(f"{machine.name} is not assigned to any production line!")
            return
        self.assigned_machine = machine
        print(f"{self.name} assigned to {machine.name} in {machine.production_line.name}")

    def operate_assigned_machine(self, product, amount):
        if not self.assigned_machine:
            print(f"{self.name} has no assigned machine!")
            return

        machine = self.assigned_machine

        if not machine.production_line:
            print(f"{machine.name} is not assigned to any production line!")
            return

        if product not in machine.production_line.products:
            print(f"{product.name} is not in {machine.production_line.name}")
            return

        if machine.status == "Maintenance":
            print(f"{machine.name} is under maintenance!")
            return

        if machine.status == "Running":
            print(f"{machine.name} is already running {machine.current_product.name}!")
            return

        print(f"{self.name} operates {machine.name}")
        machine.start(product)
        product.produce(amount)
        machine.stop()
        print(f"{self.name} finished work on {machine.name}")


class Engineer(Worker):
    def __init__(self, name, field):
        super().__init__(name)
        self.field = field

    def design(self):
        print(f"{self.name} is designing in {self.field}")


class SeniorEngineer(Engineer):
    def __init__(self, name, field, team_size):
        super().__init__(name, field)
        self.team_size = team_size

    def supervise(self):
        print(f"{self.name} supervises {self.team_size} engineers")


class FactoryManager(Worker):
    def manage_factory(self):
        print(f"{self.name} is managing the factory")


# Production Line 
class ProductionLine:
    def __init__(self, name):
        self.name = name
        self.machines = []
        self.workers = []
        self.products = []

    def add_machine(self, machine):
        machine.production_line = self
        self.machines.append(machine)

    def add_worker(self, worker):
        self.workers.append(worker)

    def add_product(self, product):
        self.products.append(product)

    def line_report(self):
        print(f"\nReport for Line: {self.name}")
        for product in self.products:
            print(f"  Product {product.name}: {product.quantity} units")

        for machine in self.machines:
            current = machine.current_product.name if machine.current_product else "None"
            print(f"  Machine {machine.name}: {machine.status}, Product: {current}")

        for worker in self.workers:
            assigned = worker.assigned_machine.name if isinstance(worker, RegularWorker) and worker.assigned_machine else "None"
            print(f"  Worker: {worker.name}, Machine: {assigned}")


# Factory
class Factory:
    def __init__(self, name):
        self.name = name
        self.production_lines = []
        self.managers = []

    def add_line(self, line):
        self.production_lines.append(line)

    def add_manager(self, manager):
        self.managers.append(manager)

    def performance_report(self):
        print(f"\nFactory: {self.name} Performance Report")
        for line in self.production_lines:
            line.line_report()
        for manager in self.managers:
            print(f"Manager: {manager.name}")


# Factory Controller 
class FactoryController:
    def __init__(self, factory):
        self.factory = factory

    # Add Production Line
    def add_production_line(self, line_name):
        line = ProductionLine(line_name)
        self.factory.add_line(line)
        print(f"Production Line '{line_name}' added.")
        return line

    # Add Machines / Products
    def add_machine_to_line(self, line, machine_name):
        machine = Machine(machine_name)
        line.add_machine(machine)
        print(f"Machine '{machine_name}' added to {line.name}")
        return machine

    def add_product_to_line(self, line, product_name):
        product = Product(product_name)
        line.add_product(product)
        print(f"Product '{product_name}' added to {line.name}")
        return product

    # Add Workers
    def add_regular_worker(self, line, worker_name):
        worker = RegularWorker(worker_name)
        line.add_worker(worker)
        print(f"Regular Worker '{worker_name}' added to {line.name}")
        return worker

    def add_engineer(self, line, name, field):
        eng = Engineer(name, field)
        line.add_worker(eng)
        print(f"Engineer '{name}' added to {line.name}")
        return eng

    def add_senior_engineer(self, line, name, field, team_size):
        seng = SeniorEngineer(name, field, team_size)
        line.add_worker(seng)
        print(f"Senior Engineer '{name}' added to {line.name}")
        return seng

    # Add Factory Manager
    def add_manager(self, name):
        manager = FactoryManager(name)
        self.factory.add_manager(manager)
        print(f"Factory Manager '{name}' added.")
        return manager

    # Assign Worker to Machine
    def assign_worker_to_machine(self, worker, machine):
        if isinstance(worker, RegularWorker):
            worker.assign_machine(machine)
        else:
            print(f"{worker.name} cannot operate machines!")

    # Operate Machine
    def operate_machine(self, worker, product, amount):
        if isinstance(worker, RegularWorker):
            worker.operate_assigned_machine(product, amount)
        else:
            print(f"{worker.name} cannot operate machines!")

    # Reports
    def production_line_report(self, line):
        line.line_report()

    def factory_report(self):
        self.factory.performance_report()
```

```python
import tkinter as tk
from tkinter import ttk, messagebox
import sqlite3

# ================== DATABASE LAYER ==================
class FactoryDB:
    def __init__(self):
        self.conn = sqlite3.connect('factory_pro_v5.db')
        self.cursor = self.conn.cursor()
        self.setup_db()

    def setup_db(self):
        self.cursor.execute('CREATE TABLE IF NOT EXISTS personnel (id INTEGER PRIMARY KEY AUTOINCREMENT, name TEXT, role TEXT, assigned_to_id INTEGER, assignment_type TEXT)')
        self.cursor.execute('CREATE TABLE IF NOT EXISTS lines (id INTEGER PRIMARY KEY AUTOINCREMENT, name TEXT)')
        self.cursor.execute('CREATE TABLE IF NOT EXISTS machines (id INTEGER PRIMARY KEY AUTOINCREMENT, name TEXT, line_id INTEGER, status TEXT DEFAULT "Idle")')
        self.cursor.execute('CREATE TABLE IF NOT EXISTS products (id INTEGER PRIMARY KEY AUTOINCREMENT, name TEXT, stock INTEGER DEFAULT 0)')
        self.conn.commit()

    def execute(self, query, params=()):
        self.cursor.execute(query, params); self.conn.commit()

    def fetch(self, query, params=()):
        self.cursor.execute(query, params); return self.cursor.fetchall()

    def vacuum_ids(self, table):
        items = self.fetch(f"SELECT * FROM {table} ORDER BY id ASC")
        self.execute(f"DELETE FROM {table}")
        self.execute(f"DELETE FROM sqlite_sequence WHERE name='{table}'")
        for item in items:
            cols = self.fetch(f"PRAGMA table_info({table})")
            col_names = [c[1] for c in cols if c[1] != 'id']
            self.execute(f"INSERT INTO {table} ({','.join(col_names)}) VALUES ({','.join(['?']*len(col_names))})", item[1:])

# ================== GUI LAYER ==================
class FactoryERP:
    def __init__(self, root):
        self.root = root
        self.root.title("FACTORY CONTROL PRO v5.0")
        self.root.state('zoomed') 
        self.db = FactoryDB()
        self.bg, self.card, self.accent = "#0A0B10", "#161B22", "#58A6FF"
        self.root.configure(bg=self.bg)
        self.setup_styles()
        self.setup_layout()

    def setup_styles(self):
        style = ttk.Style()
        style.theme_use("clam")
        style.configure("TNotebook", background=self.bg, borderwidth=0)
        style.configure("TNotebook.Tab", background=self.card, foreground="gray", padding=[30, 12], font=("Segoe UI", 12, "bold"))
        style.map("TNotebook.Tab", background=[("selected", self.accent)], foreground=[("selected", "black")])
        style.configure("Treeview", background="#0D1117", foreground="white", fieldbackground="#0D1117", rowheight=35)
        style.configure("Treeview.Heading", background="#21262D", foreground=self.accent, font=("Segoe UI", 11, "bold"))

    def setup_layout(self):
        self.sidebar = tk.Frame(self.root, width=260, bg="#010409")
        self.sidebar.pack(side="left", fill="y")
        self.sidebar.pack_propagate(False)
        tk.Label(self.sidebar, text="SYSTEM", fg=self.accent, bg="#010409", font=("Impact", 40), pady=40).pack()

        self.tabs = ttk.Notebook(self.root)
        self.tabs.pack(side="right", expand=True, fill="both", padx=15, pady=15)
        self.pages = {name: tk.Frame(self.tabs, bg=self.bg) for name in ["Dash", "Monitor", "Staff", "Assets", "Ops", "Inv"]}
        self.tabs.add(self.pages["Dash"], text=" 📊 DASHBOARD ")
        self.tabs.add(self.pages["Monitor"], text=" 🖥️ LIVE MONITOR ")
        self.tabs.add(self.pages["Staff"], text=" 👥 PERSONNEL ")
        self.tabs.add(self.pages["Assets"], text=" 🏗️ ASSETS ")
        self.tabs.add(self.pages["Ops"], text=" 🕹️ CONTROL ")
        self.tabs.add(self.pages["Inv"], text=" 📦 WAREHOUSE ")

        self.init_dash(); self.init_monitor(); self.init_staff(); self.init_assets(); self.init_inv(); self.init_ops()
        self.refresh_all()

    def init_monitor(self):
        self.mon_canvas = tk.Canvas(self.pages["Monitor"], bg=self.bg, highlightthickness=0)
        self.mon_frame = tk.Frame(self.mon_canvas, bg=self.bg)
        self.mon_scroll = ttk.Scrollbar(self.pages["Monitor"], orient="vertical", command=self.mon_canvas.yview)
        self.mon_canvas.configure(yscrollcommand=self.mon_scroll.set)
        self.mon_scroll.pack(side="right", fill="y")
        self.mon_canvas.pack(side="left", fill="both", expand=True)
        self.mon_canvas_win = self.mon_canvas.create_window((0,0), window=self.mon_frame, anchor="nw")
        self.mon_frame.bind("<Configure>", lambda e: self.mon_canvas.configure(scrollregion=self.mon_canvas.bbox("all")))
        self.mon_canvas.bind("<Configure>", lambda e: self.mon_canvas.itemconfig(self.mon_canvas_win, width=e.width))

    def update_monitor(self):
        for w in self.mon_frame.winfo_children(): w.destroy()
        lines = self.db.fetch("SELECT * FROM lines")
        for ln in lines:
            l_box = tk.Frame(self.mon_frame, bg="#161B22", highlightthickness=1, highlightbackground="#30363D")
            l_box.pack(fill="x", padx=30, pady=20)
            l_head = tk.Frame(l_box, bg="#21262D", padx=25, pady=15)
            l_head.pack(fill="x")
            tk.Label(l_head, text=f"LINE: {ln[1].upper()}", fg=self.accent, bg="#21262D", font=("Segoe UI", 22, "bold")).pack(side="left")
            
            mgr = self.db.fetch("SELECT name FROM personnel WHERE assigned_to_id=? AND role='Manager' AND assignment_type='Line'", (ln[0],))
            snr = self.db.fetch("SELECT name FROM personnel WHERE assigned_to_id=? AND role='Senior Engineer' AND assignment_type='Line'", (ln[0],))
            tk.Label(l_head, text=f"Sr. Eng: {snr[0][0]}" if snr else "Sr. Eng: VACANT", fg="#D29922", bg="#21262D", font=("Segoe UI", 12, "bold")).pack(side="right", padx=15)
            tk.Label(l_head, text=f"Manager: {mgr[0][0]}" if mgr else "Manager: VACANT", fg="#F85149", bg="#21262D", font=("Segoe UI", 12, "bold")).pack(side="right", padx=15)
            
            m_grid = tk.Frame(l_box, bg="#161B22", padx=20, pady=20); m_grid.pack(fill="x")
            machines = self.db.fetch("SELECT id, name, status FROM machines WHERE line_id=?", (ln[0],))
            for mc in machines:
                clr = "#3FB950" if mc[2]=="Running" else "#D29922" if mc[2]=="Maintenance" else "#F85149"
                mc_card = tk.Frame(m_grid, bg="#0D1117", highlightthickness=2, highlightbackground=clr, padx=15, pady=15, width=320)
                mc_card.pack(side="left", padx=10, pady=10); mc_card.pack_propagate(False); mc_card.config(height=280)
                
                tk.Label(mc_card, text=mc[1].upper(), fg="white", bg="#0D1117", font=("Segoe UI", 14, "bold")).pack(anchor="w")
                tk.Label(mc_card, text=f"● {mc[2].upper()}", fg=clr, bg="#0D1117", font=("Segoe UI", 10, "bold")).pack(anchor="w", pady=(0, 5))
                tk.Frame(mc_card, height=1, bg="#30363D").pack(fill="x", pady=5)
                
                staff_container = tk.Frame(mc_card, bg="#0D1117"); staff_container.pack(fill="both", expand=True)
                assigned_staff = self.db.fetch("SELECT name, role FROM personnel WHERE assigned_to_id=? AND assignment_type='Machine'", (mc[0],))
                if not assigned_staff:
                    tk.Label(staff_container, text="NO STAFF ASSIGNED", fg="#484F58", bg="#0D1117", font=("Segoe UI", 9, "italic")).pack(pady=20)
                else:
                    assigned_staff.sort(key=lambda x: x[1] != "Engineer")
                    for name, role in assigned_staff:
                        role_clr = "#A371F7" if role == "Engineer" else "#58A6FF"
                        s_row = tk.Frame(staff_container, bg="#161B22", pady=2, padx=5); s_row.pack(fill="x", pady=2)
                        badge = "ENG" if role == "Engineer" else "OPR"
                        tk.Label(s_row, text=badge, fg="black", bg=role_clr, font=("Segoe UI", 7, "bold"), width=4).pack(side="left", padx=(0, 5))
                        tk.Label(s_row, text=name, fg="white", bg="#161B22", font=("Segoe UI", 9), anchor="w").pack(side="left", fill="x")

    def link(self, t):
        if not self.op_u.get(): return
        uid = self.op_u.get().split(":")[0]
        role = self.db.fetch("SELECT role FROM personnel WHERE id=?", (uid,))[0][0]
        
        if t == "Machine":
            if role in ["Manager", "Senior Engineer"]:
                messagebox.showerror("Error", "Managers and Seniors cannot be assigned to Machines."); return
            target = self.op_m.get()
        else:
            if role not in ["Manager", "Senior Engineer"]:
                messagebox.showerror("Error", "Only Managers/Seniors can lead a Line."); return
            target = self.op_l.get()

        if not target: return
        tid = target.split(":")[0]
        if messagebox.askyesno("CONFIRM", "Apply assignment?"):
            self.db.execute("UPDATE personnel SET assigned_to_id=?, assignment_type=? WHERE id=?", (tid, t, uid))
            self.refresh_all()

    def set_stat(self, s):
        if not self.run_mc.get(): return
        mid = self.run_mc.get().split(":")[0]
        mname = self.run_mc.get().split(": ")[1].split(" (")[0]

        if s == "Running":
            eng_count = self.db.fetch("SELECT COUNT(*) FROM personnel WHERE assigned_to_id=? AND role='Engineer' AND assignment_type='Machine'", (mid,))[0][0]
            opr_count = self.db.fetch("SELECT COUNT(*) FROM personnel WHERE assigned_to_id=? AND role='Worker' AND assignment_type='Machine'", (mid,))[0][0]
            
            errors = []
            if eng_count == 0: errors.append(f"👤 Missing ENGINEER on '{mname}'.")
            if opr_count == 0: errors.append(f"👤 Missing WORKER on '{mname}'.")
            
            if errors:
                messagebox.showerror("Activation Failed", "Cannot start machine:\n\n" + "\n".join(errors))
                return

        if messagebox.askyesno("Confirm Status", f"Set {mname} to {s.upper()}?"):
            self.db.execute("UPDATE machines SET status=? WHERE id=?", (s, mid))
            messagebox.showinfo("Success", f"{mname} status updated to {s.upper()}.")
            self.refresh_all()

    def log_prod(self):
        if not self.p_item.get() or not self.p_qty.get().isdigit() or not self.p_line.get(): 
            messagebox.showwarning("Incomplete", "Selection missing: Product, Line, or Quantity."); return
        
        pid = self.p_item.get().split(":")[0]
        lid = self.p_line.get().split(":")[0]
        
        mgr = self.db.fetch("SELECT COUNT(*) FROM personnel WHERE assigned_to_id=? AND role='Manager' AND assignment_type='Line'", (lid,))[0][0]
        snr = self.db.fetch("SELECT COUNT(*) FROM personnel WHERE assigned_to_id=? AND role='Senior Engineer' AND assignment_type='Line'", (lid,))[0][0]
        machines = self.db.fetch("SELECT id, name, status FROM machines WHERE line_id=?", (lid,))
        
        errors = []
        if mgr == 0: errors.append("❌ LINE MANAGER is missing.")
        if snr == 0: errors.append("❌ SENIOR ENGINEER is missing.")
        if not machines: errors.append("❌ NO MACHINES registered on this line.")

        for mid, mname, status in machines:
            if status != "Running":
                errors.append(f"❌ Machine '{mname}' is {status.upper()}.")
            eng_count = self.db.fetch("SELECT COUNT(*) FROM personnel WHERE assigned_to_id=? AND role='Engineer' AND assignment_type='Machine'", (mid,))[0][0]
            opr_count = self.db.fetch("SELECT COUNT(*) FROM personnel WHERE assigned_to_id=? AND role='Worker' AND assignment_type='Machine'", (mid,))[0][0]
            if eng_count == 0: errors.append(f"👤 Machine '{mname}' requires an ENGINEER.")
            if opr_count == 0: errors.append(f"👤 Machine '{mname}' requires a WORKER.")

        if errors:
            messagebox.showerror("Production Denied", "Incomplete Line Requirements:\n\n" + "\n".join(errors))
            return
            
        if messagebox.askyesno("Confirm Batch", f"Proceed with {self.p_qty.get()} units?"):
            self.db.execute("UPDATE products SET stock = stock + ? WHERE id=?", (int(self.p_qty.get()), pid))
            messagebox.showinfo("Success", "Production successful. Inventory updated.")
            self.refresh_all()

    def init_dash(self):
        self.dash_frame = tk.Frame(self.pages["Dash"], bg=self.bg); self.dash_frame.pack(fill="both", expand=True, padx=80, pady=60)

    def update_dash(self):
        for w in self.dash_frame.winfo_children(): w.destroy()
        stats = {
            "RUNNING": (self.db.fetch("SELECT COUNT(*) FROM machines WHERE status='Running'")[0][0], "#3FB950"),
            "MAINTENANCE": (self.db.fetch("SELECT COUNT(*) FROM machines WHERE status='Maintenance'")[0][0], "#D29922"),
            "OFFLINE": (self.db.fetch("SELECT COUNT(*) FROM machines WHERE status='Idle'")[0][0], "#F85149"),
            "STAFF": (self.db.fetch("SELECT COUNT(*) FROM personnel")[0][0], "#58A6FF"),
            "LINES": (self.db.fetch("SELECT COUNT(*) FROM lines")[0][0], "#A371F7"),
            "STOCK": (self.db.fetch("SELECT SUM(stock) FROM products")[0][0] or 0, "#8B949E")
        }
        for i, (label, (val, color)) in enumerate(stats.items()):
            f = tk.Frame(self.dash_frame, bg=self.card, highlightthickness=2, highlightbackground=color, padx=30, pady=30)
            f.grid(row=i//3, column=i%3, padx=15, pady=15, sticky="nsew")
            tk.Label(f, text=label, fg="gray", bg=self.card, font=("Segoe UI", 14, "bold")).pack()
            tk.Label(f, text=str(val), fg=color, bg=self.card, font=("Arial", 50, "bold")).pack(pady=10)
        for i in range(3): self.dash_frame.columnconfigure(i, weight=1)

    def init_ops(self):
        f1 = tk.LabelFrame(self.pages["Ops"], text=" 🔗 ASSIGNMENT ", fg=self.accent, bg=self.bg, pady=15); f1.pack(fill="x", padx=20, pady=10)
        self.op_u = ttk.Combobox(f1, width=35); self.op_u.grid(row=0, column=0, padx=20)
        self.op_m = ttk.Combobox(f1, width=20); self.op_m.grid(row=0, column=1, padx=10)
        tk.Button(f1, text="TO MACHINE", bg="#1F6FEB", fg="white", command=lambda: self.link("Machine")).grid(row=0, column=2, padx=5)
        self.op_l = ttk.Combobox(f1, width=20); self.op_l.grid(row=0, column=3, padx=10)
        tk.Button(f1, text="TO LINE", bg="#1F6FEB", fg="white", command=lambda: self.link("Line")).grid(row=0, column=4, padx=5)
        
        f2 = tk.LabelFrame(self.pages["Ops"], text=" ⚡ STATUS CONTROL ", fg=self.accent, bg=self.bg, pady=15); f2.pack(fill="x", padx=20, pady=10)
        self.run_mc = ttk.Combobox(f2, width=35); self.run_mc.pack(side="left", padx=20)
        for s, c in [("Running", "#238636"), ("Maintenance", "#D29922"), ("Idle", "#F85149")]:
            tk.Button(f2, text=s.upper(), bg=c, fg="white", width=14, command=lambda x=s: self.set_stat(x)).pack(side="left", padx=10)
        
        f3 = tk.LabelFrame(self.pages["Ops"], text=" 📦 PRODUCTION LOG ", fg="#3FB950", bg="#0D1117", pady=20); f3.pack(fill="x", padx=20, pady=15)
        self.p_item = ttk.Combobox(f3, width=20); self.p_item.pack(side="left", padx=5)
        self.p_line = ttk.Combobox(f3, width=15); self.p_line.pack(side="left", padx=5)
        self.p_qty = tk.Entry(f3, width=8, font=("Arial", 12, "bold")); self.p_qty.pack(side="left", padx=5)
        tk.Button(f3, text="START BATCH", bg="#238636", fg="white", command=self.log_prod).pack(side="left", padx=20)

    def init_staff(self):
        f = tk.Frame(self.pages["Staff"], bg=self.card, pady=10); f.pack(fill="x", padx=20)
        self.st_n = tk.Entry(f); self.st_n.pack(side="left", padx=5)
        self.st_r = ttk.Combobox(f, values=["Worker", "Engineer", "Senior Engineer", "Manager"]); self.st_r.pack(side="left", padx=5); self.st_r.current(0)
        tk.Button(f, text="Add", command=lambda: [self.db.execute("INSERT INTO personnel (name, role) VALUES (?,?)", (self.st_n.get(), self.st_r.get())), self.refresh_all()]).pack(side="left")
        self.st_tree = ttk.Treeview(self.pages["Staff"], columns=(1,2,3), show="headings"); [self.st_tree.heading(i, text=h) for i,h in enumerate(["ID","Name","Role"],1)]; self.st_tree.pack(fill="both", expand=True, padx=20)
        tk.Button(self.pages["Staff"], text="Delete Selected", bg="#F85149", fg="white", command=lambda: self.del_it("personnel", self.st_tree)).pack(pady=10)

    def init_assets(self):
        f1 = tk.LabelFrame(self.pages["Assets"], text=" Lines ", fg=self.accent, bg=self.bg, pady=10); f1.pack(fill="both", expand=True, padx=20, pady=5)
        self.ln_e = tk.Entry(f1); self.ln_e.pack(side="left", padx=10)
        tk.Button(f1, text="Add Line", command=lambda: [self.db.execute("INSERT INTO lines (name) VALUES (?)", (self.ln_e.get(),)), self.refresh_all()]).pack(side="left")
        self.ln_tree = ttk.Treeview(f1, columns=(1,2), show="headings", height=3); [self.ln_tree.heading(i, text=h) for i,h in enumerate(["ID","Name"],1)]; self.ln_tree.pack(side="left", fill="both", expand=True, padx=10)
        tk.Button(f1, text="Delete", bg="#F85149", fg="white", command=lambda: self.del_it("lines", self.ln_tree)).pack(side="right", padx=10)

        f2 = tk.LabelFrame(self.pages["Assets"], text=" Machines ", fg=self.accent, bg=self.bg, pady=10); f2.pack(fill="both", expand=True, padx=20, pady=5)
        self.mc_e = tk.Entry(f2); self.mc_e.pack(side="left", padx=10)
        self.mc_l = ttk.Combobox(f2); self.mc_l.pack(side="left", padx=5)
        tk.Button(f2, text="Add Machine", command=lambda: [self.db.execute("INSERT INTO machines (name, line_id) VALUES (?,?)", (self.mc_e.get(), self.mc_l.get().split(":")[0])), self.refresh_all()]).pack(side="left", padx=10)
        self.mc_tree = ttk.Treeview(f2, columns=(1,2,3), show="headings", height=5); [self.mc_tree.heading(i, text=h) for i,h in enumerate(["ID","Machine","Status"],1)]; self.mc_tree.pack(side="left", fill="both", expand=True, padx=10)
        tk.Button(f2, text="Delete", bg="#F85149", fg="white", command=lambda: self.del_it("machines", self.mc_tree)).pack(side="right", padx=10)

    def init_inv(self):
        f = tk.Frame(self.pages["Inv"], bg=self.card, pady=10); f.pack(fill="x", padx=20)
        self.pr_n = tk.Entry(f); self.pr_n.pack(side="left", padx=10)
        tk.Button(f, text="Add Product", command=lambda: [self.db.execute("INSERT INTO products (name) VALUES (?)", (self.pr_n.get(),)), self.refresh_all()]).pack(side="left")
        self.pr_tree = ttk.Treeview(self.pages["Inv"], columns=(1,2,3), show="headings"); [self.pr_tree.heading(i, text=h) for i,h in enumerate(["ID","Product","Stock"],1)]; self.pr_tree.pack(fill="both", expand=True, padx=20)
        tk.Button(self.pages["Inv"], text="Delete Item", bg="#F85149", fg="white", command=lambda: self.del_it("products", self.pr_tree)).pack(pady=10)

    def del_it(self, tbl, tree):
        sel = tree.selection()
        if sel and messagebox.askyesno("Confirm Delete", "Permanently remove this record?"):
            self.db.execute(f"DELETE FROM {tbl} WHERE id=?", (tree.item(sel[0])['values'][0],))
            self.db.vacuum_ids(tbl); self.refresh_all()

    def refresh_all(self):
        self.update_dash(); self.update_monitor()
        for t, q in [(self.st_tree, "SELECT id, name, role FROM personnel"), (self.pr_tree, "SELECT id, name, stock FROM products"), (self.ln_tree, "SELECT id, name FROM lines"), (self.mc_tree, "SELECT id, name, status FROM machines")]:
            for i in t.get_children(): t.delete(i)
            for r in self.db.fetch(q): t.insert("", "end", values=r)
        self.op_u['values'] = [f"{x[0]}: {x[1]} ({x[2]})" for x in self.db.fetch("SELECT id, name, role FROM personnel")]
        self.op_m['values'] = [f"{x[0]}: {x[1]}" for x in self.db.fetch("SELECT id, name FROM machines")]
        self.op_l['values'] = [f"{x[0]}: {x[1]}" for x in self.db.fetch("SELECT id, name FROM lines")]
        self.mc_l['values'] = [f"{x[0]}: {x[1]}" for x in self.db.fetch("SELECT id, name FROM lines")]
        self.run_mc['values'] = [f"{x[0]}: {x[1]} ({x[2]})" for x in self.db.fetch("SELECT id, name, status FROM machines")]
        self.p_item['values'] = [f"{x[0]}: {x[1]}" for x in self.db.fetch("SELECT id, name FROM products")]
        self.p_line['values'] = [f"{x[0]}: {x[1]}" for x in self.db.fetch("SELECT id, name FROM lines")]

if __name__ == "__main__":
    try:
        from ctypes import windll
        windll.shcore.SetProcessDpiAwareness(1)
    except: pass 
    root = tk.Tk(); app = FactoryERP(root); root.mainloop()
```
